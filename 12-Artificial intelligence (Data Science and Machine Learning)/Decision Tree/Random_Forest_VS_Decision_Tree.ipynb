```python
from sklearn.datasets import load_iris
from sklearn import tree
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
X, Y = iris.data, iris.target
inputs = pd.DataFrame(X, columns=iris.feature_names)
outputs = Y
```

```python
X_train , X_test, Y_train, Y_test = train_test_split(inputs,outputs,test_size = 0.2)
```

```python
X_train
```

```python
DT = tree.DecisionTreeClassifier()
DT = DT.fit(X_train, Y_train)
```

```python
print("accuracy: ",DT.score(X_test,Y_test))
```

```python
tree.plot_tree(DT)
```

# In scikit-learn, the binary search algorithm is used to find the optimal threshold value for each continuous feature in a decision tree.

```python
import graphviz 
dot_data = tree.export_graphviz(DT, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("DT_CLASS") 
```

```python
dot_data = tree.export_graphviz(DT, out_file=None,feature_names=iris.feature_names,class_names=iris.target_names,filled=True, rounded=True,special_characters=True)  
graph = graphviz.Source(dot_data)  
graph
```

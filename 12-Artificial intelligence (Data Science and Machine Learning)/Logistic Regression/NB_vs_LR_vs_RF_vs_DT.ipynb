```python
import seaborn as sns
import pandas as pd 
```

```python
df = pd.read_csv("/content/Training set.csv")
df.head()
```

```python
sns.scatterplot(x="Height", y="Weight", data=df, hue="Sex",palette = 'YlGnBu')
```

```python
for i in range(len(df.Sex)):
  if df.Height[i] > 400 or df.Weight[i] > 200: 
    df = df.drop(index  = i)
```

```python
sns.scatterplot(x="Height", y="Weight", data=df, hue="Sex",palette = 'YlGnBu')
```

```python
X = df[["Height","Weight"]]
Y = df["Sex"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
DT = DecisionTreeClassifier()
DT.fit(X_train,Y_train)
DT.score(X_test,Y_test)
```

```python
RF = RandomForestClassifier(n_estimators=1000)
RF.fit(X_train,Y_train)
RF.score(X_test,Y_test)
```

```python
LR = LogisticRegression()
LR.fit(X_train,Y_train)
LR.score(X_test,Y_test)
```

```python
NB = GaussianNB()
NB.fit(X_train,Y_train)
NB.score(X_test,Y_test)
```

#### We see that Naive Bayes and Logistic Regression are the most performant algorithms, compared to Random Forest, which requires a significant amount of time to train, and Decision Tree, which gives us a lower score compared to the other algorithms.

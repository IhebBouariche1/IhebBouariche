```python
import pandas as pd 
import numpy as np 
```

```python
Data = pd.read_csv("./tweet_emotions.csv")
Data
```

```python
Data.shape
```

```python
emotions =[]
for x in Data.sentiment:
  emotions.append(x)
emotions = np.array(emotions)
print(emotions)
```

```python
content =[]
for x in Data.content:
  content.append(x)
content = np.array(content)
print(content)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(content)
```

```python
#words index and print it 
word_index = tokenizer.word_index
#Generate sequences of the texts
sequences = tokenizer.texts_to_sequences(content)
#padding the sequences to a uniform lenght 
pad = pad_sequences(sequences, padding='post' )             #post: ignore the lasts words , pre : igrore the firsts 
print(word_index)
```

```python
print(pad.shape)
```

```python
tokenizer_output = Tokenizer()
tokenizer_output.fit_on_texts(emotions)
#words index and print it 
word_index_outputs = tokenizer_output.word_index
#Generate sequences of the texts
sequences = tokenizer_output.texts_to_sequences(emotions)
#padding the sequences to a uniform lenght 
pad_output = pad_sequences(sequences, padding='post' )             #post: ignore the lasts words , pre : igrore the firsts 
print(word_index_outputs)
print(pad_output.shape)
```

```python
output = np.zeros((40000,13))
for i in range(40000):
  x=pad_output[i]
  output[i,x-1]=1
print(pad_output)
print("\n")
print(output)
```

```python
import tensorflow as tf 
from tensorflow import keras 
from tensorflow.keras import layers

#Build the model
model = keras.Sequential([
    layers.Embedding(48998,64,input_length=37),
    layers.Bidirectional(layers.LSTM(64,return_sequences=True)),
    layers.Bidirectional(layers.LSTM(64)),
    layers.Dense(128,activation='relu'),
    layers.Dense(13,activation='softmax')
])
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",
              optimizer=keras.optimizers.Adam(lr=0.0001),
              metrics=["accuracy"]
     )
```

```python
model.fit(pad,output,epochs=30)
```

```python
test = ['i hate when you love others']
 

test_sequences = tokenizer.texts_to_sequences(test)
#padding the sequences to a uniform lenght 
test = pad_sequences(test_sequences, padding='post' , maxlen = 37 )             #post: ignore the lasts words , pre : igrore the firsts 

print(test.shape)
print(test)
```

```python
model.predict(test)
```

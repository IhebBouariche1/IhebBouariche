```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
# We will choose one of these data and work on it
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size='size',palette = 'YlGnBu')
```

```python
sns.histplot(tips['tip'],kde=True)
```

```python
sns.barplot(x="sex", y="tip",data=tips ,palette = "YlGnBu")
```

```python
sns.boxplot(x="day", y="total_bill",data=tips ,palette = "YlGnBu")
```

```python
sns.stripplot(x="day", y="total_bill",hue = "sex" , data=tips ,palette = "YlGnBu")
```

```python
sns.jointplot(x='tip',y='total_bill', data=tips, kind="reg")
```

```python
sns.jointplot(x='tip',y='total_bill', data=tips, kind="kde", shade = True, cmap = "YlGnBu")
```

```python
sns.jointplot(x='tip',y='total_bill', data=tips, kind="hex", cmap = "YlGnBu")
```

```python
# We will work now on titanic dataset
titanic = sns.load_dataset('titanic')
titanic 
```

```python
sns.pairplot(titanic.select_dtypes(['number']),hue = "pclass")
```

```python
#titanic data correlation (correlation measures the strength and direction of the linear relationship between two variables.)
titanic.corr()
```

```python
# with these graphs we can interpret data very easily 
sns.heatmap(titanic.corr(), annot = True, cmap = "YlGnBu") 
```
